```python
def find_begin_end_disease(id):
    i = 0
    begin = 0
    end = len(doc.paragraphs)
    for paragraph in doc.paragraphs:
        if id in paragraph.text:
            begin = i
        number = int(id.split('/T')[1]) + 1
        if "ZYYXH/T"+str(number) in paragraph.text:
            end = i
        i += 1
    return begin, end
```

```python
def gather_aim_disease_info(name, id):
    begin, end = find_begin_end_disease(id)
    find = False
    dic = {'id':[],'疾病':[],'证候名称':[],'证候':[],'治法':[],'formulas':[],'成分':[]}
    concate = False
    add_name = ""
    for i in range(begin, end):
        if "其他治法" in doc.paragraphs[i].text:
            break
        if "辨证论治" in doc.paragraphs[i].text:
            find = True
            continue
        if find:
            if "2\ue010" in doc.paragraphs[i].text:
                if len(dic['证候']) < len(dic['证候名称']):
                    dic['证候名称'].pop()
                if "2\ue010" in doc.paragraphs[i+1].text:
                    add_name = doc.paragraphs[i].text.split('\ue010')[-1]
                    concate = True
                else:
                    if concate and len(doc.paragraphs[i].text.split('\ue010')) > 2:
                        dic['证候名称'].append(add_name + "-" + doc.paragraphs[i].text.split('\ue010')[-1])
                    else:
                        dic['证候名称'].append(doc.paragraphs[i].text.split('\ue010')[-1])
            elif "证候：" in doc.paragraphs[i].text:
                dic['id'].append(id)
                dic['疾病'].append(name)
                dic['证候'].append(doc.paragraphs[i].text.split('：')[1])
            elif "治法：" in doc.paragraphs[i].text or '治疗：' in doc.paragraphs[i].text:
                dic['治法'].append(doc.paragraphs[i].text.split('：')[1])
            elif "方药：" in doc.paragraphs[i].text:
                dic['formulas'].append(doc.paragraphs[i].text.split('：')[1])
                dic['成分'].append(doc.paragraphs[i+1].text)
    if len(dic['证候']) < len(dic['证候名称']):
        dic['证候名称'].pop()
    print(dic)
    return dic
```

```python
def get_result(list_disease):
    dic0 = pd.DataFrame()
    false = ""
    unsecessful_list = []
    for i in list_disease:
        try:
            false = i
            result = gather_aim_disease_info(i,dic_id_disease[i])
            dic1 = pd.DataFrame(result)
            dic0 = pd.concat([dic0,dic1])
            print(i,"sucessful")
        except Exception:
            unsecessful_list.append(false)
    return dic0, unsecessful_list
```

```python
list_disease = ['肥胖','高脂血症', '心力衰竭','冠心病心绞痛','室性早搏','病毒性心肌炎','高血压','普通感冒','慢性呼吸衰竭','慢性胃炎', '肠易激综合征', '慢性病毒性肝炎', '肝硬化', '慢性肾衰竭' , '类风湿性关节炎', '焦虑症', '抑郁症', '阿尔茨海默病']
result = get_result(list_disease)
result
```

# Specification of drug names

```python
result_all = pd.read_excel('all_disease.xlsx')
```

```python
drug = result_all[['formulas','成分']]
```

```python
drug.reset_index(inplace=True)
```

```python
drug
drug['成分'] = drug['成分'].str.replace('；', '，')
```

```python
split_result = {'formulas':[],'成分':[], '中药名称':[], '剂量':[]}
```

```python
import re
pattern = r'(\D+)(\d+\D+)'
error_formular = []
for i in range(len(drug)):
    formulas = drug.loc[i,'formulas']
    ingredient = drug.loc[i,'成分']
    try:
        for j in drug.iloc[i,2].split('，'):      
            match1 = re.match(pattern, j)
            TRM_name = match1.group(1)
            dose = match1.group(2)
            split_result['formulas'].append(formulas)
            split_result['成分'].append(ingredient)
            split_result['中药名称'].append(TRM_name)
            split_result['剂量'].append(dose)
    except:
        error_formular.append(formulas)
```

```python
split_result = pd.DataFrame(split_result)
```

```python
split_result['剂量'] = split_result['剂量'].str.replace('。', '')
split_result['formulas'] = split_result['formulas'].str.replace('。', '')
split_result['成分'] = split_result['成分'].str.replace('。', '')
split_result['中药名称'] = split_result['中药名称'].str.replace('后下', '')
split_result['中药名称'] = split_result['中药名称'].str.replace('先煎', '')
```

```python

result1 = split_result.loc[split_result['中药名称'].str.contains('、')]
```
