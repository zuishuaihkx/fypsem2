```python
def hanzi_to_pinyin(text):
    pinyin_list = pinyin(text, style=Style.NORMAL)
    pinyin_text = ' '.join([item[0] for item in pinyin_list])
    return pinyin_text
```

```python
import requests
import json

def get_consitituent(name):
# URL of the API
    url = 'http://batman2api.cloudna.cn/queryTarget'

    # POST request parameters as described in the image
    payload = {
        "content": [
            {
                "clusterName": "jing jie",
                "type": "caoyao",
                "list": [name]
            }
        ],
        "pvalue": 0.05,
        "userInScore": 1000
    }

    # Headers to specify the type of data being sent
    headers = {
        'Content-Type': 'application/json'
    }

    # Make the POST request
    response = requests.post(url, data=json.dumps(payload), headers=headers)

    # Check if the request was successful
    if response.status_code == 200:
        # Process the response if necessary
        # For example, print the response text
        data = response.json()
        
        # If the data is a list of records, directly convert it to a DataFrame
        if isinstance(data, list):
            df = pd.DataFrame(data)
        # If the data is a dictionary, you might need to normalize it before converting
        else:
            df = pd.json_normalize(data)
        
        # Now df is a pandas DataFrame containing the response data
        print(name + ': successful')
        return df
    else:
        print(f"Error: {response.status_code} {name}")
        return None
        
```
